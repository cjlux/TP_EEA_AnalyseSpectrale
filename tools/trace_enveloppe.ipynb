```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def weight(t_ech, t1, n):
    T = []
    for t in t_ech:
        if t < t1: 
            T.append(t/t1)
        else:
            T.append(np.exp(t1-t)**n)
    return np.array(T)
```

```python
Fs = 44100
Ts = 1/Fs
D  = 2
t_ech = np.arange(0, D+Ts, Ts)
```

```python
len(t_ech)
```

```python
plt.figure()
plt.plot(t_ech, weight(t_ech, 20e-3, 3))
```
